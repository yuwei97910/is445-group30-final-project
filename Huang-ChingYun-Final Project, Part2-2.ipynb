```python
select_box = alt.binding_select(options=list(df_IL['stars'].unique()))
selection = alt.selection_single(name='x_axis', fields=['stars'], bind=select_box)

ys = ['stars', 'review_count', 'is_open', 'latitude', 'longitude']

select_box = alt.binding_select(options=ys, name='y')
sel_y = alt.selection_single(fields=['y'], bind=select_box, init={'y': 'latitude'})

xs = ['stars', 'review_count', 'is_open', 'latitude', 'longitude']

select_box = alt.binding_select(options=xs, name='x')
sel_x = alt.selection_single(fields=['x'], bind=select_box, init={'x': 'longitude'})


alt.Chart(df_IL).transform_fold(
    ys,
    as_ = ['y', 'y_value']
).transform_fold(
    xs,
    as_ = ['x', 'x_value']
).mark_point().encode(
    x='x_value:Q',
    y='y_value:Q',
    tooltip='name'
).add_selection(
    sel_y,
    sel_x,
    selection
).transform_filter(
    sel_y
).transform_filter(
    sel_x
).transform_filter(
    selection
).interactive()
```
