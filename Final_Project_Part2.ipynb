```python
def dashboard(state):
    df_state = df_IL
    if state == 'FL':
        df_state = df_FL
    elif state == 'IN':
        df_state = df_IN
    elif state == 'PA':
        df_state = df_PA
#     select_box = alt.binding_select(options=list(df_state['stars'].unique()))
#     selection = alt.selection_single(name='x_axis', fields=['stars'], bind=select_box)

    ys = ['is_open', 'latitude', 'longitude','Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday', 'has_wifi','has_parking', 'animal_friendly']

    select_box = alt.binding_select(options=ys, name='y_axis')
    sel_y = alt.selection_single(fields=['y_axis'], bind=select_box, init={'y_axis': 'Monday'})

    xs = ['stars', 'review_count']

    select_box = alt.binding_select(options=xs, name='x_axis')
    sel_x = alt.selection_single(fields=['x_axis'], bind=select_box, init={'x_axis': 'stars'})


    chart = alt.Chart(df_IL).transform_fold(
        ys,
        as_ = ['y_axis', 'y_value']
    ).transform_fold(
        xs,
        as_ = ['x_axis', 'x_value']
    ).mark_point().encode(
        x='x_value:Q',
        y='y_value:Q',
        tooltip='name',
        color=alt.Color('x_value:Q')
    ).add_selection(
        sel_y,
        sel_x,
    ).transform_filter(
        sel_y
    ).transform_filter(
        sel_x
    ).interactive()
    chart.display()
```

```python
state_list = ['IL', 'FL', 'IN','PA']
```

```python
widgets.interact(dashboard, state = state_list)
```

### Write up

For this assignment, we created a graph with drop-down options for user to select the column they want and review the relationships between each columns. In the plot, users can first select a state, which we subset the data from the master dataset "yelp_academic_dataset_business", and then look into the relationship of different columns. In the drop-down list, we have five different variables related to the restaurants' information - "stars", "review_count", "is_open", "latitude", and "longitude." 

Originally, we aimed to use two different datasets - "yelp_academic_dataset_business" and "yelp_academic_dataset_review." But after our discussion, we decided to mainly use "yelp_academic_dataset_business" dataset as we currently subset the dataset into four different datasets based on states to meet the file hosting requirement of Github. We may include more states in our final dashboard. We will take "yelp_academic_dataset_review" as contextual dataset as we may include attributes such as review stars count of reach stores. "yelp_academic_dataset_review" dataset can also be found on https://www.yelp.com/dataset/download

We found that it's a little bit hard to ensure the label can changed with the different selections of dropdown list. So this time, we decide not to change the label. But we will fix this problem in final work using JavaScript.
